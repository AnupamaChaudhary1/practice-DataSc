```python
import seaborn as sns
import matplotlib.pyplot as plt
p=sns.get_dataset_names()
print(p)
```

```python
# load datasets
data=sns.load_dataset('iris')
print(data)
print(data.columns)
plt.figure(figsize=(3,2))
sns.lineplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Title using matplotlib')
plt.show()
```

```python
sns.set_style('dark')
data=sns.load_dataset('iris')
sns.lineplot(x='sepal_length',y='species',data=data)
#setting the title using matplotlib
plt.title('Title using matplotlib function')
plt.show()
```

```python
plt.figure(figsize=(3,2))
sns.barplot(x='species',y='sepal_width',data=data)
plt.title('Bar graph')
plt.show()
```

```python
sns.countplot(x='species',data=data)
plt.title('count plot for species')
plt.show()
```

```python
sns.set_style('dark')
sns.pairplot(data=data, hue='sepal_width')
plt.show()
```
