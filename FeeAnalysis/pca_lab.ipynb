```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Simulate student scores with more features
np.random.seed(42)
data=pd.DataFrame({
    'maths':np.random.randint(30,100,100),
    'science':np.random.randint(30,100,100),
    'english':np.random.randint(30,100,100),
    'attendance':np.random.randint(30,100,100),
    'homework':np.random.randint(30,100,100)
})

print(data.head(5))
#Standardize the data
scaler=StandardScaler()
data_scaled=scaler.fit_transform(data)

print(data_scaled[:5])

#apply PCA
pca=PCA(n_components=2)
data_pca=pca.fit_transform(data_scaled)
print(data_pca)

#analyze
loadings=pd.DataFrame(pca.components_.T,columns=['PC1','PC2'],index=data.columns) #execute result
print(loadings)
#visualize
plt.scatter(data_pca[:,0],data_pca[:,1],alpha=0.6)
plt.xlabel('Principle component 1')
plt.ylabel('Principle component 2')
plt.title('PCA projection of Student Data')
plt.grid(True)
plt.show()
```
