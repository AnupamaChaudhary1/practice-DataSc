```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import joblib



```

```python
#sample data
data={
    'text': [
        'This is good movie!'," It was fantastic film","Really enjoyable","Awsome action",
        'Worst movie ever', "I hated the movie", "Terrible ","Bad acting"
    ],
    'label': [1,1,1,1,0,0,0,0]
}
```

```python
#processing and train
df=pd.DataFrame(data)
X=df['text']
y=df['label']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

vectorizer=CountVectorizer()
X_train_vec=vectorizer.fit_transform(X_train)

model=LinearRegression()
model.fit(X_train_vec,y_train)
```

```python
#save model and vactorizer
joblib.dump(model,'sentiment_model.pkl')
joblib.dump(vectorizer,'vectorizer.pkl')
```

```python
# 2. Load cleaned dataset
df = pd.read_csv("../cleaned_data.csv")
# View the first few rows
df.head()
```

```python
# 4. Create binary target variable
# Let's define 'sentiment' based on Fee Increase % (YoY): Above median = Negative (high), Below = Positive (low)
median_fee_increase = df['Fee Increase % (YoY)'].median()
df['sentiment'] = (df['Fee Increase % (YoY)'] <= median_fee_increase).astype(int)
```

```python
# 5. Select only features related to facilities, fee, and performance
features = ['Admission Fee (NPR)', 'Infrastructure Score', 'Average Academic Score (%)']

X = df[features]
y = df['sentiment']

# Scale the selected features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 6. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# 7. Train model
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# 8. Save model and scaler
joblib.dump(model, "school_sentiment_model.pkl")
joblib.dump(scaler, "feature_scaler.pkl")
```

```python
# 9. Make prediction on new data
# Example input: [admission_fee, infra_score, academic_score]
sample_input = np.array([[15000, 80, 85]])  # only 3 features

# Scale the input with the previously fitted scaler
sample_scaled = scaler.transform(sample_input)

# Predict
prediction = model.predict(sample_scaled)[0]
sentiment = "Positive" if prediction == 1 else "Negative"
print("Predicted Sentiment:", sentiment)
```
