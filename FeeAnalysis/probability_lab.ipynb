#Probability Distributions
- It describes how the values of a random variable are distributed.
- it shows all the possible values a variable can take and the probability of each value occuring.
eg: tossing a coin (Head or tail)
 rolling a dice
 scores of students in a class(normally distributed)


types of distributions
1. Uniform distribution
 all outcomes are equally likely.
 - eg: rolling a fair dice, flipping a coin

```python
import numpy as np
import matplotlib.pyplot as plt
outcomes= np.arange(1,7)
print(outcomes)
probability=[1/6]*6
print(probability)

plt.bar(outcomes,probability)
plt.title('Uniform distribution')
plt.xlabel('outcomes')
plt.ylabel('probability')
plt.show()
```

2. Bernoulli Distribution
 - only two outcomes possible: either success or failure (0)
  eg: tossing a coin, answering yes no qs, pass/ fial in exam

```python
import seaborn as sns
from scipy.stats import bernoulli
data=bernoulli.rvs(p=0.5, size=1000)
'''' 
p=0.5 : this means probability of success is 50%
size=1000: we flipping the coin 1000 times.
'''
# print(data)
sns.histplot(data,kde=False,bins=2)
plt.title('bernoulli distribution')
plt.xlabel('outcomes')
plt.ylabel('Frequency')
plt.show()
```

3. Binomial Distribution
 - repeated bernoulli trials
 - A binomial distribution models the number of success in a fixed number of independent trials, each with some probabillity of success.
eg: probability of getting heads in 10 coin tosses.
 

```python
from scipy.stats import binom
import numpy as np
n=10
p=0.5
x=np.arange(0,n+1)
y= binom.pmf(x,n,p) # probability mass function
print(y)

plt.bar(x,y)
plt.title('Binomial Distribution (n=10,p=0.5)')
plt.xlabel('Number of Success')
plt.ylabel('probability')
plt.show()
```

4. Normal Distribution (Gaussian)
 - Bell shaped Curve
 - most commonly used in real world eg; height, weight, test scores
 - most values cluster around the mean (center) and fewer are found as we move away.
 

```python
from scipy.stats import norm
import numpy as np
data=np.random.randint(35,100,size=100)
mean=np.mean(data)
std=np.std(data)
print(mean,std)
x=np.linspace(mean-3*std, mean+3*std, 100)
# generates 100 evenly spaced values between m-3std,m+3std
# print(x)
y=norm.pdf(x,mean,std)
plt.plot(x,y)
plt.title('Normal distribution')
plt.xlabel('Values')
plt.ylabel('Probability distribution')
plt.grid()
plt.show()
```

5. Poisson Distribution
 - used to model count based model in a fixed interval
 eg: no of emails received in an hour
     number of cars passing by a checkpoint in an hour
     number of typing errors per page
     

```python
from scipy.stats import poisson
x=np.arange(0,15)
#possible number of events (eg, 0 cars, 1 cars, 2 cars ....in a minute)
print(x)
# average number of events
lamb=20
pmf=poisson.pmf(x, lamb)
plt.bar(x, pmf)
plt.xlabel('Number of events')
plt.ylabel('probability')
plt.grid()
plt.show()
```

Hypotheisi testing:  Imagine we are curious 
- do boys and irls perform equally in maths?
- We cannot ask every student in Nepal so, we collect sample data and make a guess, this is called hypothesis.
- we then test whether our guess is supported or not using math logic.
- >>>> Hypothesis
- > null hypotheis (H0): Nothing is differentm things are as expected.
- eg boys abd girls perform equally in math

- > Alternative Hypothesi (H1): sth is different
- boys and girls perform differently in maths

- Eg.: is the new drug better than the old one?
- did the new teaching method improve performance?
- do users prefer older iphones over new iphones?

-  Steps to test the hypotheis
1. Make a guess: H0 or H1
2. choose a risk level (sigma: usually 0.05)
3. get t-sscore,z-score
4. check how rare our result is p-value
5. decide: if rare:reject H0
