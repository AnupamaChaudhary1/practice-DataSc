```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,25,30,40]
# Set figure size  - line chart
plt.figure(figsize=(3,2))
plt.plot(x,y, color='green',linewidth='3', marker='.', markerfacecolor='blue', markersize=12, linestyle='dashed')
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)   # grids means small boxes
plt.show()

#Bar chart
plt.figure(figsize=(3,2))
subjects=['Maths','Science','English']
scores=[80,90,75]
plt.bar(subjects,scores)
plt.title('Subject scores')
plt.ylabel("Marks")
plt.show()
```

```python
#Subplots
fig,axis=plt.subplots(1,2, figsize=(8,3)) # in 1 row show two figures
axis[0].bar(subjects,scores)

axis[1].plot(x,y,color='green',linewidth='3', marker='.', markerfacecolor='blue', markersize=12, linestyle='dashed')
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)   # grids means small boxes
plt.show()
plt.suptitle("Two charts side by side")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('cleaned_data.csv') 
```

```python
# Calculate average values
average_fees = {
    'Annual Tuition Fee': df['Annual Tuition Fee (NPR)'].mean(),
    'Admission Fee': df['Admission Fee (NPR)'].mean(),
    'Monthly Fee': df['Monthly Fee (NPR)'].mean()
}

# Plot bar graph
plt.figure(figsize=(5,3))
plt.bar((average_fees.keys()), (average_fees.values()), color='skyblue')
# sns.barplot(x=list(average_fees.keys()), y=list(average_fees.values()), palette="Blues_d")
plt.title("Average Fees Comparison")
plt.ylabel("Amount (NPR)")
plt.xlabel("Fee Type")
plt.show()
```

```python
# Sort by Fee Increase %
sorted_df = df.sort_values('Fee Increase % (YoY)')

plt.figure(figsize=(10, 6))
plt.plot(sorted_df['Fee Increase % (YoY)'], sorted_df['Average Academic Score (%)']) # , color='purple', alpha=0.7
plt.title("Fee Increase % vs Academic Score")
plt.xlabel("Fee Increase % (YoY)")
plt.ylabel("Average Academic Score (%)")
plt.grid(True)
plt.show()
```

```python
# Prepare average values
averages = {
    'Academic Score (%)': df['Average Academic Score (%)'].mean(),
    'Technology Access Index': df['Technology Access Index'].mean(),
    'Infrastructure Score': df['Infrastructure Score'].mean()
}

plt.figure(figsize=(7,5))
plt.bar((averages.keys()), (averages.values()), color='green')
# sns.barplot(x=list(averages.keys()), y=list(averages.values()), palette="Greens_d")
plt.title("Average Academic & Infrastructure Scores")
plt.ylabel("Score / Index")
plt.xlabel("Category")
plt.show()
```

```python
left=[1,2,3,4,5]
height=[10,20,30,40,50]
#lable
tick_label=['one','two','three','four','five']
#plotting
plt.bar(left,height,tick_label=tick_label,width=0.8,color=['b','g','y','grey','pink'],label=tick_label)
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar graph')
plt.show()
```
