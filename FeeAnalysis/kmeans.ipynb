```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#simulate
np.random.seed(42)
data=pd.DataFrame({
    'maths':np.random.randint(30,100,100),
    'science':np.random.randint(30,100,100)
})

#Apply kmeans
kmeans=KMeans(n_clusters=3,random_state=42)
kmeans.fit(data)

#add cluster label
data['cluster']=kmeans.labels_

# Get cluster centers
centroids=kmeans.cluster_centers_

# print clusters
print(data)

print(centroids)
"""
[100 rows x 3 columns]
[[78.44827586 46.10344828]
 [40.87179487 61.02564103]
 [75.34375    85.84375   ]]

 cluster 0: 78.44827586 46.10344828 > High in maths, but low in science
 cluster 1: 40.87179487 61.02564103 > Weaker in maths, but average in science
 cluster 2: 75.34375    85.84375 > High in both maths and science.
"""

#visualize
plt.figure(figsize=(8,6))
scatter=plt.scatter(data['maths'],data['science'],c=data['cluster'],label=data['cluster'])
plt.scatter(centroids[:,0],centroids[:,1],c='red',s=150,marker='x',label='Centroids')

plt.xlabel('Maths Score')
plt.ylabel('Science Score')
plt.title('KMeans clustering of students marks')
plt.legend(*scatter.legend_elements())
plt.grid(True)
plt.show()
```
